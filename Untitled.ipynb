# 実践：データの操作

## 表を表示させる

```python
import pandas as pd
iris = pd.read_csv('iris.data.csv',  names=['sepal_length[cm]', 'sepal_width[cm]', 
                                            'petal_length[cm]', 'petal_width[cm]', 'class'])
# sepal_length : がく片の長さ,   sepal_width: がく片の幅
# petal_length: 花びらの長さ,    petal_width: 花びらの幅
```

```python
iris.head(150)
```

classには、'setosa', 'virginica', 'versicolor'の3種類の品種が登録されています。
<br>このデータを用いてグラフ化を行います。

## グラフ化させてみる

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(iris, hue="class")
```
